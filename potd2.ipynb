```python
'''
If:

n = 3

Then the square has:

3 rows
3 columns

Visual form:


***
* *
***



if n = 1,
the output should be [*] because 1 row and 1 column
'''
```

```python
def generate_hollow_square(n):
    star = '*'
    l = []
    for i in range(n):
        for j in range(n):
            if (i == 0) or (i== n-1) or (j==0) or (j == n-1):
                print(star,end='')
            else:
                print(" ",end='')
        print()
```

```python
x = generate_hollow_square(3)
print(x)
```
